```python
def break_by_month(line):
    df = ttc_station_df[ttc_station_df['Line'] == line]
    group_one = df.groupby(['Station','Date']).sum()
    group_two = group_one.unstack().T
    group_two.index = group_two.index.get_level_values(1)
    return group_two.resample('M').sum()
```

```python
def cc(arg):
    '''
    Shorthand to convert 'named' colors to rgba format at 60% opacity.
    '''
    return mcolors.to_rgba(arg, alpha=0.3)
```

```python
def polygon_under_graph(xlist, ylist):
    '''
    Construct the vertex list which defines the polygon filling the space under
    the (xlist, ylist) line graph.  Assumes the xs are in ascending order.
    '''
    return [(xlist[0], 0.)] + list(zip(xlist, ylist)) + [(xlist[-1],0)]
```

```python
import matplotlib.dates as mdates
def plot_3d_line(df, groups, ylabels, *args, **kwargs):
    c = kwargs.get('c', None)
    x_axis = kwargs.get('x_axis', None)
    x_label = kwargs.get('x_label', None)
    
    xlabel = kwargs.get('xlabel', None)
    ylabel = kwargs.get('ylabel', None)
    zlabel = kwargs.get('zlabel', None)
    maxValue = kwargs.get('maxValue', None)
    
    fig = plt.figure(figsize=(12,9))
    ax = fig.gca(projection='3d')
    
    x_num = df.shape[0]
    y_num = df.shape[1]
    
    index = 0
    verts = []
    
    for g in groups:
        ys = df[g].values
        x = x_axis
        verts.append(polygon_under_graph(x, ys))
    
    
    facecolors = []
    for color in c:
        
        facecolors = facecolors + [cc(color)]
        

    
    poly = PolyCollection(verts, facecolors=facecolors)

    
    ax.add_collection3d(poly, zs=np.arange(0, y_num), zdir='y')
    
    
    ax.set_xticks(np.arange(0, len(x_label), 4))
    ax.set_xticklabels(x_label)
    
    ax.set_yticks(np.arange(0,y_num))
    ax.set_yticklabels(ylabels, rotation=-15)
    
#     ax.view_init(0, 30)

    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_zlabel(zlabel)
    ax.set_xlim(0, x_num)
    ax.set_ylim(0, y_num)
    ax.set_zlim(0, 600)
    
    ax.xaxis.set_major_formatter(FuncFormatter(lambda x, _ : pd.to_datetime(x_label[x]).strftime('%m-%Y')))
    
```

```python
stations = top_ten_delays_yongedf.columns.values
times = top_ten_delays_yongedf.index.values


number_index = np.arange(0, len(top_ten_delays_yongedf.index))

test = top_ten_delays_yongedf.copy()
test.index = number_index
test

colors = ['#113344'] * 2 + ['r'] + ['#113344'] * 7

ax = plot_3d_line(test, stations, stations,c =colors, x_axis=number_index, x_label=times, ylabel='Station', xlabel='Time', zlabel='Sum Delays (min)')

# top10_delays_sum_yongedf_fig = plt.gcf()
# top10_delays_sum_yongedf_fig.title('Sum of Delays on Yonge-University line')
# plt.gcf()
plt.title('Top 10 Stations with the most delays on YU line')
plt.savefig('Top ten delays YU.png')
plt.show()

# plt.title('Sum of Delays on Yonge-University line')
# plt.savefig('Sum of Delays on Yonge-University line.png')
# plt.show()
```

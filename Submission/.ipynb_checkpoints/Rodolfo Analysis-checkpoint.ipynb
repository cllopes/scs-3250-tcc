```python
import pandas as pd
from pandas import DataFrame
from pylab import *
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from dateutil.parser import parse
%matplotlib inline
```

```python
from pandas.plotting import autocorrelation_plot
```

```python
ttc = pd.read_csv('ttc_delays_cleaned.csv', header = 0,  
                          encoding='utf8' , sep=",")
```

```python
ttc.head()
```

```python
ttc.tail()
```

```python
# Convert Date to datetime object
ttc['Date'] = pd.to_datetime(ttc['Date'])
```

```python
ttc.tail()
```

```python
ttc.index = ttc['Date'] ## Set the datetime column as the index, I created a new index, and did not convert the date indo 
```

```python
ttc.head()
```

```python
ttc_month = ttc.resample('M')
```

```python
ttc_month['Min Delay'].plot()
```

```python
ttc_month_count = ttc.resample('M').count().drop('Vehicle', 1)
```

```python
ttc_month_count.head()
```

```python
ttc_month_count_plot = ttc_month_count['Min Delay'].plot()
ttc_month_count_plot.set_ylabel("Event counts")
ttc_month_count_fig = ttc_month_count_plot.get_figure()
ttc_month_count_fig.tight_layout()
ttc_month_count_fig.savefig('Monthly event count.png')
```

```python
ttc_month_count_autocorrelation_plot = autocorrelation_plot(ttc_month_count)
ttc_month_count_autocorrelation_fig = ttc_month_count_autocorrelation_plot.get_figure()
ttc_month_count_autocorrelation_fig.tight_layout()
ttc_month_count_autocorrelation_fig.savefig('Autocorrelation Monthly event count.png')
```

```python
ttc_month_sum = ttc.resample('M').sum().drop('Vehicle', 1)
```

```python
ttc_month_sum.head()
```

```python
ttc_month_sum_plot = ttc_month_sum['Min Delay'].plot()
ttc_month_sum_plot.set_ylabel("Sum of Delays (min)")
ttc_month_sum_fig = ttc_month_sum_plot.get_figure()
ttc_month_sum_fig.tight_layout()
ttc_month_sum_fig.savefig('Monthly sum of delays.png')
```

```python
autocorrelation_plot(ttc_month_sum)
```

```python
ttc_month_mean = ttc.resample('M').mean().drop('Vehicle', 1)
```

```python
ttc_month_mean.head()
```

```python
ttc_month_mean_plot = ttc_month_mean['Min Delay'].plot()
ttc_month_mean_plot.set_ylabel("Mean of Delays (min)")
ttc_month_mean_fig = ttc_month_mean_plot.get_figure()
ttc_month_mean_fig.tight_layout()
ttc_month_mean_fig.savefig('Monthly mean of delays.png')
```

Lower mean delay in July might correspond to Pan AM games that took place in Toronto

```python
autocorrelation_plot(ttc_month_mean)
```

Horizontal lines displayed in the plot correspond to 95% and 99% confidence bands. Therefore, we can say that the mean delay times are strongly correlated going back ten months.

```python
ttc.index.names = ['Datetime']
```

```python
ttc_bound_pivot = pd.pivot_table(ttc, index=["Date","Bound"],
               values=["Min Delay"],
               fill_value=0,margins=True).reset_index()
```

```python
ttc_bound_pivot.head()
```

```python
ttc_bound_pivot['Date'] = pd.to_datetime(ttc_bound_pivot['Date'].astype(str), errors='coerce')
ttc_bound_pivot.head()
```

```python
ttc_bound_df = ttc_bound_pivot.set_index('Date')
ttc_bound_df.head()
```

```python
East = ttc_bound_df[ttc_bound_df['Bound']=='East']
West = ttc_bound_df[ttc_bound_df['Bound']=='West']
North = ttc_bound_df[ttc_bound_df['Bound']=='North']
South = ttc_bound_df[ttc_bound_df['Bound']=='South']
```

```python
East.head()
```

```python
East_month_sum = East.resample('M').sum()
West_month_sum = West.resample('M').sum()
North_month_sum = North.resample('M').sum()
South_month_sum = South.resample('M').sum()
```

```python
East_month_sum.head()
```

```python
fig, axs = plt.subplots(1,4, figsize = (15,5))

East_month_sum.plot(ax=axs[0]).set_title('East')
West_month_sum.plot(ax=axs[1]).set_title('West')
North_month_sum.plot(ax=axs[2]).set_title('North')
South_month_sum.plot(ax=axs[3]).set_title('South')
```

```python
#autocorrelation_plot(East_month_sum)
autocorrelation_plot(West_month_sum)
#autocorrelation_plot(North_month_sum)
#autocorrelation_plot(South_month_sum)
```

```python
East_month_mean = East.resample('M').mean()
West_month_mean = West.resample('M').mean()
North_month_mean = North.resample('M').mean()
South_month_mean = South.resample('M').mean()
```

```python
East_month_mean.head()
```

```python
fig, axs = plt.subplots(1,4, figsize = (15,5))

East_month_mean.plot(ax=axs[0]).set_title('East')
West_month_mean.plot(ax=axs[1]).set_title('West')
North_month_mean.plot(ax=axs[2]).set_title('North')
South_month_mean.plot(ax=axs[3]).set_title('South')
axs[0].set_ylabel('Mean delay time (min)')
axs[1].set_ylabel('Mean delay time (min)')
axs[2].set_ylabel('Mean delay time (min)')
axs[3].set_ylabel('Mean delay time (min)')
fig.tight_layout()

Mean_delay_time_per_direction = plt.gcf()
Mean_delay_time_per_direction.savefig('Mean_delay_time_per_direction.png')
plt.show()
```

```python
autocorrelation_plot(East_month_mean)
autocorrelation_plot(West_month_mean)
autocorrelation_plot(North_month_mean)
autocorrelation_plot(South_month_mean)
```
